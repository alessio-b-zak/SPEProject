```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.spatial.distance as distance
import scipy.cluster.hierarchy as hierarchy
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
```

```python
data_load = np.genfromtxt(fname='determinand-demographic.csv', delimiter=',', dtype=None)
data_values = data_load[1:, 3::2].astype(np.int)

data_load_names = pd.read_csv('determinand-demographic-names.csv')
data_names = data_load_names.values[1:]

data = np.append(data_names, data_values, 1)
# dtype = [('name', 'S10'), ('2000', int), ('2001', int), ('2002', int), ('2003', int), ('2004', int), ('2005', int), ('2006', int), ('2007', int), ('2008', int), ('2009', int), ('2010', int), ('2011', int), ('2012', int), ('2013', int), ('2014', int), ('2015', int)]
# data = np.array(data,dtype=dtype)
```

```python
def plot_top_30(matrix):
    rows = 10
    cols = 3
    pylab.rcParams['figure.figsize'] = (70.0, 120.0)
    fig = plt.figure()
    plotID = 1
    years = []
    for year in range(2000, 2016):
        years.append(year)
    for i in range(rows):
        for j in range(cols):
            ax = fig.add_subplot( rows, cols, plotID )
            ax.plot(years, matrix[plotID - 1, 1:])
            ax.set_title(matrix[plotID - 1, 0])
            ax.set_ylim(0, 200000)
            plotID += 1
    # fig.suptitle('Training Data Scatter Plot Matrix', fontsize=60)
    plt.show()
```

```python
plot_top_30(data)
```
